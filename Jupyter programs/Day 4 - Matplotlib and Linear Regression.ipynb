```python
import pandas as pd
dataset=pd.read_csv('C:/Users/Nandan/Dropbox/PG-DBDA CDAC Mumbai/ML/Data.csv')
```

```python
#splitting the dataset into dependent class label Y and the independent feature set X
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
dataset_conv = pd.get_dummies(dataset, columns=['Country'])
X = np.append(dataset_conv.iloc[:,3:6].values,dataset_conv.iloc[:,:2].values,axis=1)
y = dataset_conv.iloc[:,2].values
X[:,3:5] = imputer.fit_transform(X[:,3:5])

X
y
```

```python
#splitting the dataset into training and testing sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)
```
