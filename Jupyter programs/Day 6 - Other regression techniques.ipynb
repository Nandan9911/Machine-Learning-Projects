```python
import pandas as pd
import numpy as pn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#upload scikit-learn in-built boston housing dataset
from sklearn import datasets
boston=datasets.load_boston()
boston
```

```python
print(boston.DESCR) #in built parameter to give the description of the dataset: features description and source of origin of dataset
```

```python
boston.target
```

```python
bostonDF=pd.DataFrame(boston.data,columns=boston.feature_names)
bostonDF.head()
```

```python
boston
```

```python
sns.pairplot(bostonDF)
```

```python
plt.show()
```

```python
bostonDF['House Price']=boston.target
```

```python
bostonDF.shape
```

```python
X = bostonDF.iloc[:,0:13]
y = bostonDF.iloc[:,13]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
#Preprocessing

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
lm_coeff=pd.Series(lm.coef_,index=boston.feature_names)
lm_coeff
# shows the coefficients for each and every feature
```

```python
lm_coeff.plot(kind="bar")
```

```python
# The loss is calculated from the 1st expression (loss fucntion)
# Loss is the difference between actual and predicted values respectively
# To minimize the loss, we use the regularization term
# We calculate the regularization using Rigde and Lasso Regression
```

# Ridge regression

```python
from sklearn.linear_model import Ridge
rid=Ridge()
rid.fit(X_train,y_train)
```

```python
rid_coeff=pd.Series(rid.coef_,index=boston.feature_names)
rid_coeff
```

```python
# As the value of lambda is changed (increased, decreased), the coefficients of the model are also changed, and thus the efficiency of the model can be checked
# if the dataset is big, and the coefficient(s) of some parameter(s) are also large, them the computation cost is high in terms of time
# Hence, we are using Lasso, Ridge regression to change the coefficients and check the best fit
```

```python
rid_coeff.plot(kind="bar")
plt.show()
```
