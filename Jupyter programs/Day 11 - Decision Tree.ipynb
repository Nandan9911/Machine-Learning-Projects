```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
sn = pd.read_csv('C:/Users/Nandan/Dropbox/PG-DBDA CDAC Mumbai/ML/Social_Network_Ads.csv')
```

```python
sn
```

```python
X=sn.iloc[:,:-1].values
y=sn.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#Decision Tree Algorithm

from sklearn.tree import DecisionTreeClassifier #the gaussian distribution is usually called the normal distribution
model=DecisionTreeClassifier(criterion='entropy', random_state=0)
model2=model.fit(X_train,y_train)
model2
```

```python
print(model2.predict(sc.transform([[42,98000]])))
```

```python
y_pred=model2.predict(X_test)
y_pred
```

```python
np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)), axis=1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```
