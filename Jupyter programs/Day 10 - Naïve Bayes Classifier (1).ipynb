```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
sn = pd.read_csv('C:/Users/Nandan/Dropbox/PG-DBDA CDAC Mumbai/ML/Social_Network_Ads.csv')
```

```python
sn
```

```python
X=sn.iloc[:,:-1].values
y=sn.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#Naïve Bayes Algorithm
#GaussianNB --> Gaussian distribution for naive bayes
from sklearn.naive_bayes import GaussianNB # the gaussian distribution is usually called the normal distribution
model=GaussianNB()
model2=model.fit(X_train,y_train)
model2
```

```python
print(model2.predict(sc.transform([[28,95000]])))
```

```python
y_pred=model2.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```
